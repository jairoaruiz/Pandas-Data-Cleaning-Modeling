```python
import pandas as pd
import statsmodels.formula.api as smf
```

```python
acs = pd.read_csv('./acs_ny.csv')
acs.head()
```

```python
results = smf.poisson('NumChildren ~ FamilyIncome + FamilyType + OwnRent', data=acs).fit()
```

```python
results.summary2()
```

```python
import statsmodels
import statsmodels.api as sm
```

###  GLM

```python
model = smf.glm('NumChildren ~ FamilyIncome + FamilyType + OwnRent', data=acs, 
                family=sm.families.Poisson(sm.genmod.families.links.log))
```

```python
results = model.fit()
results.summary2()
```

```python
model = smf.glm('NumChildren ~ FamilyIncome + FamilyType + OwnRent', data=acs, 
                family=sm.families.NegativeBinomial(sm.genmod.families.links.log))
results = model.fit()
```

```python
results.summary2()
```
